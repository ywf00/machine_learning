```python
import numpy as np
import nnfs
from nnfs.datasets import spiral_data

nnfs.init()
np.random.seed(0)


class layer:
    def __init__(self,n_inputs,n_nuerons):
        self.weights = 0.1*np.random.randn(n_inputs,n_nuerons)
        self.biases = np.zeros((1,n_nuerons))
    def forward(self,inputs):
        self.output = np.dot(inputs,self.weights)+self.biases


class ReLU:
    def forward(self,inputs):
        self.output = np.maximum(0,inputs)



class softmax:
    def forward(self,input):
        expo = np.exp(input- np.max(input,axis=1,keepdims=True))
        norm = expo/np.sum(expo,axis=1,keepdims=True)
        self.output = norm


class loss:  
    def calculate(self,output,y):
        sample_losses = self.forward(output, y)
        data_loss = np.mean(sample_losses)
        return data_loss
    


class loss_cce(loss):    
    def forward(self,y_pred,y_true):
        samples = len(y_pred)
        y_pred_clip = np.clip(y_pred,1e-7, 1-1e-7)


        if len(y_true.shape) == 1:
            correct = y_pred_clip[range(samples),y_true]
        elif len(y_true.shape)== 2:
            correct = np.sum(y_pred_clip * y_true,axis=1)  

        neglog = -np.log(correct)
        return neglog      


X,y = spiral_data(samples=100, classes=3)
l1 = layer(2,3)
relu = ReLU()
l2 = layer(3,3)
soft = softmax()
lossfun = loss_cce()
los = lossfun.calculate(soft,y)

l1.forward(X)
relu.forward(l1.output)
l2.forward(relu.output)
soft.forward(l2.output)


print("Loss:",los)
```
